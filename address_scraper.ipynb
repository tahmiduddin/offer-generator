```python
import requests

from bs4 import BeautifulSoup
import pandas as pd
import time

# URL of the website to scrape
url = "https://bsaonline.com/OnlinePayment/OnlinePaymentSearchResults?PaymentSearchCategory=Address&PaymentApplicationType=Tax&PaymentSearchText=600-1000%20E%20Barrett&uid=269&DetailResultsGrid-size=15&DetailResultsGrid-page=4"
```

```python
# Send an HTTP GET request to the website
response = requests.get(url)
# Parse the HTML code using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
addresses = soup.find_all('td')
for address in addresses:
  print(address.text)
```
